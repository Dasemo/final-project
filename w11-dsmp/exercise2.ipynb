```python
class Rectangle:
    def __init__(self, base, height, square = False):
        self.base, self.height, self.square = base, height, square
        if self.base == self.height:
            square = True

    @property
    def base(self) -> int:
        return self.__base
    
    @base.setter
    def base(self, base:int):
        if type(base) != int:
            raise TypeError("Base must be a number.")
        elif base < 0:
            return ValueError("Base must be a number greater than 0.")
        else:
            self.__base = base

    @property
    def height(self) -> int:
        return self.__height
    
    @height.setter
    def height(self, height:int):
        if type(height) != int:
            raise TypeError("Height must be a number.")
        elif height < 0:
            raise ValueError("Height must be a number greater tha 0.")
        else:
            self.__height = height
    
    def __str__(self):
        if self.square == True:
            shape = 'square'
        else:
            shape = 'rectangle'

        return f"a {shape} of {self.base} x {self.height}"
    
    def __repr__(self):
        return self.__str__()
    
    def __eq__(self, another):
        return (self.height == another.height
                and self.base == another.base)
    
    def area(self):
        return(self.base * self.height)
    
    def perimeter(self):
        return(2 *self.base + 2 *self.height)
    
    def biggest_side(self):
        return max(self.base, self.height)
    
    def make_square(self):
        biggest = self.biggest_side()
        if biggest == self.height:
            self.height = biggest
            return True
        else:
            self.height = biggest
            return False
```

```python
import random

rectangles = []
for i in range(random.randint(10, 1000)):
    rectangles.append(Rectangle(random.randint(1, 10), random.randint(1, 10)))

biggest_area = max(rectangles, key = lambda r: r.area())
biggest_perimeter = max(rectangles, key = lambda r: r.perimeter())
biggest_side = max(rectangles, key = lambda r: r.biggest_side())

for i in rectangles:
    if i.base == i.height:
        print(i)

print(f"\nBiggest area: {biggest_area}")
print(f"Biggest perimeter: {biggest_perimeter}")
print(f"Biggest side: {biggest_side}")

is_base_changed = biggest_side.make_square()

if is_base_changed == True:
    attribute = 'base'
else:
    attribute = 'height'

print(f"After making the biggest side square: {biggest_side} the {attribute} was changed")
```
